## Problema de las cartas
En este problemas se tiene diez cartas numeradas del 1 al 10. Se necesita encontrar dos subconjuntos de cartas disyuntos, donde la suma sea 36 y el producto sea 360. Un ejemplo visual de este problema se muestra a continuacion:

image.png

Este problema se puede plantear como un problema de parametros binarios, donde las cartas que tengan asignado 1 (o True) sean utilizadas para evaluar la suma, mientras que las cartas restantes (con valor 0 o False) son utilizadas para evaluar el producto.

```python
from numpy import round, around, zeros, random, append
```

```python
#Definicion de la funcion de aptitud
#Problema de las cartas
def evalbin(a):
    suma=0
    prod=1
    for i in range(a.size):
        if a[i]:
            suma+=i+1
        else:
            prod*=i+1
            
    return 10*(suma==36)+10*(prod==360)
```

```python
#Genetico con codificacion binaria
#numero de iteraciones
NI=1000
#numero de cromosomas
nc=100
#numero de genes
ng=10
#problacion inicial
p=around(random.rand(nc,ng)) #matriz de unos y ceros
#probabilidad de cruce
pc=0.9
#probabilidad de mutacion
pm=0.5
fa=zeros(nc)
#evaluar poblacion inicial
for k in range(nc):
     fa[k]=evalbin(p[k])


flag=0
for k in range(NI):
    #Seleccionar padres
    p1=int(round(random.rand()*(nc-1)))
    p2=int(round(random.rand()*(nc-1)))
    #cruzar padres
    if random.rand()<=pc:
        punto=int(round((ng-1)*random.rand()))
        hijo1=append(p[p1,:punto], p[p2,punto:])
        hijo2=append(p[p2,:punto], p[p1,punto:])
    #mutar hijos
    if random.rand()<=pm:
        pos=int(round(random.rand()*(ng-1)))
        hijo1[pos]=1.0-hijo1[pos]
        pos=int(round(random.rand()*(ng-1)))
        hijo2[pos]=1.0-hijo2[pos]

    #evaluar hijos
    hijo1e=evalbin(hijo1)
    hijo2e=evalbin(hijo2)
    #insertar hijos en la poblacion
    if hijo1e>fa[p1]:
        p[p1]=hijo1
        fa[p1]=hijo1e

    if hijo2e>fa[p2]:
        p[p2]=hijo2
        fa[p2]=hijo2e
```

# Problema de la Mochila

El objetivo del problema es maximizar el benefinicio sin superar la capacidad. En general estos problemas se puede definir a partir de las siguiente ecuaciones: <br>
$n:$ numero de objetos. $a_j:$ peso de cada objeto $j$. $c_j:$ utilidad de cada objeto $j$. $b:$ la capcidad de la mochila.

$$ x_j = \left\{ \begin{matrix} 1 & \text{si el objeto $j$ se mete en la mochila.} \\ 0 & \text{si no se mete.} \end{matrix}\right.
$$

La función de aptitud esta dada como:
$$ f(\mathbf{x}) = \sum_{j=1}^{n} c_j x_j \leq b$$

La condición de capacidad esta dada como:
$$ \sum_{j=1}^{n} a_j x_j \leq b$$
