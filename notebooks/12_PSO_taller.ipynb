# Taller Optimizaci√≥n Meta-Heuristica basada en Enjambre de Particulas (PSO)

```python
import numpy as np
```

```python
fobj = lambda x: (x[0]-1)**2+(x[1]-2)**2
#Para 2<=x[0]<=4 , 1<=x[1]<=3
```

Parametros para la configuracion del PSO

```python
#PSO
iterMax = 100 #Cantidad de iteraciones del algoritmo
N = 50 #Cantidad particulas (individuos)
dim = 2 #Cantidad de parametros por particula
c1 = 0.5
c2 = 0.5
```

Limites inferiores y superiores de cada parametro

```python
linf = np.array([2,1])
lsup = np.array([4,3])
```

Construccion de la poblacion inicial, de la forma mas lenta, usando ciclos for de Ptyhon.

```python
X = np.zeros((N,dim))
for n in range(N):
  for par in range(dim):
    X[n,par] = (lsup[par]-linf[par])*np.random.rand()+linf[par]
```

Construir el vector $F$ de la evaluacion de cada particula

```python
F = np.zeros(N)
for n in range(N):
  F[n] = fobj(X[n,:])
```

Extraemos el mejor global $g_b$, que corresponde a la particula con el valor de la funcion objetivo mas peque√±o.

```python
gb = X[F.argmin(),:]
```

Hacemos una copia de $X$ para representar el mejor local de cada particula

```python
Pb = X.copy()
Fb = F.copy()
```

```python
V = np.zeros((N,dim)) #inicializar matriz de velocidades
for iter in range(iterMax):
  for n in range(N):
    for par in range(dim):
      V[n,par] = c1*np.random.rand()*(Pb[n,par]-X[n,par])+c2*np.random.rand()*(gb[par]-X[n,par])
      X[n,par] = X[n,par] + V[n,par]
    F[n] = fobj(X[n,:]) #evaluar particula
    if F[n]<Fb[n]:
      Pb[n,:] = X[n,:] #Actualizar mejor local
      Fb[n] = F[n] #Actualizar valor del mejor local
  gb = X[F.argmin(),:] #Actualizar mejor global
```
