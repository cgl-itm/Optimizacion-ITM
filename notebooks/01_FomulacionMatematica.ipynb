# Ejemplo formulación matemática de problemas de optimización

Un forma general de presentar un problema de optimizacion a partir de ecuaciones matematicas, se muestra a continuacion

$$
\begin{array}{ll}
& \min _{\boldsymbol{x}} f(\boldsymbol{x}) \\
\text { s.t. } & c_i(\boldsymbol{x})=0, \forall i \in E \\
& c_i(\boldsymbol{x}) \leq 0, \forall i \in I
\end{array}
$$

Graficar el conjunto faltible de soluciones para

$$
\begin{array}{ll}
\min_x & \left(x_1-2\right)^2+\left(x_2-1\right)^2 \\
\text { s.t. } & x_2-2 x_1=0 \\
& x_1^2-x_2 \leq 0 \\
& x_1+x_2 \leq 2
\end{array}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0,3,100)
x2 = np.linspace(-2,3,100)
x1v, x2v = np.meshgrid(x1, x2, indexing='ij')
```

```python
objfun = lambda x1,x2: (x1-2)**2+(x2-1)**2
ce = lambda x1: 2*x1
cd1 = lambda x1: x1**2 #x2 >= x1^2
cd2 = lambda x1: 2-x1 #x2 >= 2-x1
```

```python
fobj = objfun(x1v,x2v)
fce = ce(x1)
fcd1 = cd1(x1)
fcd2 = cd2(x1)
```

```python
fig,ax=plt.subplots(1,1)
CS = ax.contour(x1v, x2v, fobj,levels=[0.1,0.4,1,1.9,2.6], cmap=plt.cm.bone)
ax.clabel(CS, inline=True, fontsize=10)
ax.plot(x1,fce)
ind = fcd1<4.
ax.plot(x1[ind],fcd1[ind])
ax.plot(x1,fcd2)
ax.set_xlim([0,3])
ax.set_ylim([0,3])
```

```python
fig.savefig('ax2_figure.png')
```
